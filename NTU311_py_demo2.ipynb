```python
import time
now = time.time()
my_df = pd.read_csv('./data/fb_{}.csv'.format(my_input))
my_df = my_df.drop([0, 1, 2 ,3])
my_df["置放天數"] = (my_df["時間數值"] - now) / 86400
x_arr = my_df["置放天數"]
y_arr = my_df["按讚"]
my_df
```

```python
plt.scatter(x_arr,y_arr)
```

```python
from sklearn.model_selection import train_test_split
train_df, validation_df = train_test_split(my_df, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
X_train = train_df["置放天數"].values.reshape(-1, 1)
Y_train = train_df["按讚"].values.reshape(-1, 1)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
theta_0 = regressor.intercept_[0]
theta_1 = regressor.coef_[0,0]
```

```python
x_arr = my_df["置放天數"]
y_arr = my_df["按讚"]
plt.scatter(x_arr,y_arr)
```

```python
x_max = 0
x_min = np.min(x_arr)
x2_arr = np.linspace(x_min, x_max, 100)
y2_arr = theta_0 + theta_1 * x2_arr
```

```python
plt.scatter(x_arr,y_arr)
plt.plot(x2_arr,y2_arr, linewidth=3, color="red", alpha=0.6)
plt.show()
```
