```python
#將需要的東西import
from pyquery import PyQuery as pq
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time

#用selenium開啟chrome
driver = webdriver.Chrome("./chromedriver")

#用selenuim開啟威秀查詢場次的網站
driver.get("https://www.vscinemas.com.tw/ShowTimes/")
```

```python
#用selenium控制下拉式選單選擇想要爬的影城
select = Select(driver.find_element_by_name('CinemaNameTWInfoF'))
select.select_by_value('TP')
time.sleep(2)
```

```python
html = driver.find_element_by_css_selector('*').get_attribute('outerHTML')
doc = pq(html)
driver.close()
```

```python
dataList=[]
for eachItem in doc("#ShowTimeSecondInfo>div").items():
    dataDict = {}
    dataDict["title"] = eachItem(".LangTW.MovieName").text()
    dataDict["date"] = {}
    for i,j in zip(eachItem(".LangTW.RealShowDate").items(),eachItem(".SessionTimeInfo").items()):
        dataDict["date"]["{}".format(i.text())] = j.text().split()
    dataList.append(dataDict)
dataList
```
