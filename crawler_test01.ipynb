# read HTML file

```python
from pyquery import PyQuery as pq
import codecs
f=codecs.open("my_PTT_post.html", 'r')
html = f.read()
```

```python
doc = pq(html)
for i in doc("#main-content").items():
    print(i.text())
```

```python
for i in doc("#main-content").items():
    for j in (i(".push").items()):
        print(j(".push-tag").text(),end="\t")
        print("{:12s}".format(j(".push-userid").text()),end=" ")
        print("{:<30s}".format(j(".push-content").text()))
        print(j(".push-ipdatetime").text(),end="\n")
```
